```python
import pandas as pd
```

```python
usa_previous = (51425 * 0.09959754603367016) + 51425
usa_latest = 53889
nj_latest = 72093
nj_previous = (68981 * 0.09959754603367016) + 68981
```

```python
#Median Household Income
median_incomelatest = pd.read_csv("median_income_1115.csv")
median_incomeprev = pd.read_csv("median_income_0509.csv")
```

```python
median_incomelatest.head(5)
```

```python
median_incomelatest = median_incomelatest.rename(columns = {'GEO.display-label':'GEOdisplaylabel'})
median_incomeprev = median_incomeprev.rename(columns = {'GEO.display-label':'GEOdisplaylabel'})
```

```python
median_incomelatest = median_incomelatest[median_incomelatest['GEO.id'] != 'Id']
median_incomeprev = median_incomeprev[median_incomeprev['GEO.id'] != 'Id']
```

```python
median_incomelatest['GEOdisplaylabel'] = median_incomelatest['GEOdisplaylabel'].replace(to_replace=', New Jersey', value='', regex=True)
median_incomeprev['GEOdisplaylabel'] = median_incomeprev['GEOdisplaylabel'].replace(to_replace=', New Jersey', value='', regex=True)
```

```python
median_incomelatest.head(5)
```

```python
median_incomelatest_newcols = pd.DataFrame(median_incomelatest.GEOdisplaylabel.str.split(',',1).tolist(), columns = ['GEOdisplaylabel', 'County'])
```

```python
median_incomelatest = pd.concat([median_incomelatest_newcols, median_incomelatest], axis=1)
```

```python
median_incomelatest['County'] = median_incomelatest['County'].replace(to_replace='County', value='', regex=True)
```

```python
median_incomelatest.columns = ['GEOdisplaylabel', 'County', 'GEO.id', 'GEO.id2', 'GEOdisplaylabel_long', 'median_income_latest', 'median_income_latest_moe']
median_incomelatest.head(5)
```

```python
median_incomeprev_newcols = pd.DataFrame(median_incomeprev.GEOdisplaylabel.str.split(',',1).tolist(), columns = ['GEOdisplaylabel', 'County'])
median_incomeprev = pd.concat([median_incomeprev_newcols, median_incomeprev], axis=1)
median_incomeprev['County'] = median_incomeprev['County'].replace(to_replace='County', value='', regex=True)
median_incomeprev.columns = ['GEOdisplaylabel', 'County', 'GEO.id', 'GEO.id2', 'GEOdisplaylabel_long', 'median_income_prev_woinf', 'median_income_prev_moe_inf']
median_incomeprev.head(5)
```

```python
#Merge Household Income
median_income_merge = pd.merge(median_incomelatest, median_incomeprev, left_on='GEO.id2', right_on='GEO.id2', how='left')
median_income_merge.head(5)
```

```python
len(median_incomelatest.index)
```

```python
median_income_merge = median_income_merge.drop(['GEOdisplaylabel_y', 'County_y', 'GEO.id_y', 'GEOdisplaylabel_long_y'], axis=1)
median_income_merge.rename(columns={'GEOdisplaylabel_x': 'GEOdisplaylabel', 'County_x': 'County', 'GEO.id_x': 'GEO.id', 'GEOdisplaylabel_long_x': 'GEOdisplaylabel_long'}, inplace=True)
```

```python
median_income_merge.head(5)
```

```python
median_income_merge.dtypes
```

```python
#Inflation Rate
inflation_rate = ((252.185-229.343)/229.343)
```

```python
inflation_rate
```

```python
#Add inflation
median_income_merge['median_income_prev'] = (pd.to_numeric(median_income_merge['median_income_prev_woinf'], errors='coerce') * inflation_rate) + pd.to_numeric(median_income_merge['median_income_prev_woinf'], errors='coerce') 
median_income_merge.head(5)
```

```python
median_income_merge['median_income_moe_prev'] = (pd.to_numeric(median_income_merge['median_income_prev_moe_inf'], errors='coerce') * inflation_rate) + pd.to_numeric(median_income_merge['median_income_prev_moe_inf'], errors='coerce') 
median_income_merge.head(5)
```

```python
median_income_merge.dtypes
```

```python
#median_income_merge['median_income_change'] = ((pd.to_numeric(median_income_merge['median_income_latest'])-pd.to_numeric(median_income_merge['median_income_prev'], errors='coerce'))/pd.to_numeric(median_income_merge['median_income_prev'], errors='coerce'))*100
#median_income_merge['nj_median_income_change'] = ((pd.to_numeric(median_income_merge['nj_median_income_latest'])-pd.to_numeric(median_income_merge['nj_median_income_prev'], errors='coerce'))/pd.to_numeric(median_income_merge['nj_median_income_prev'], errors='coerce'))*100
#median_income_merge['usa_median_income_change'] = ((pd.to_numeric(median_income_merge['usa_median_income_latest'])-pd.to_numeric(median_income_merge['usa_median_income_prev'], errors='coerce'))/pd.to_numeric(median_income_merge['usa_median_income_prev'], errors='coerce'))*100
```

```python
#Take out County subdivisions not defined
median_income_merge = median_income_merge[median_income_merge['GEOdisplaylabel'] != "County subdivisions not defined"]
```

```python
median_income_merge.head(4)
```

```python
#Find MOE percent
median_income_merge['median_income_moe_latest_per'] = (pd.to_numeric(median_income_merge['median_income_latest_moe'], errors='coerce')/pd.to_numeric(median_income_merge['median_income_latest'], errors='coerce'))*100 
```

```python
median_income_merge['median_income_moe_prev_per'] = (pd.to_numeric(median_income_merge['median_income_moe_prev'], errors='coerce')/pd.to_numeric(median_income_merge['median_income_prev'], errors='coerce'))*100 
```

```python
median_income_merge.head(4)
```

```python
#Add NJ and Nationwide
median_income_merge['nj_median_income_latest'] = nj_latest
median_income_merge['nj_median_income_prev'] = nj_previous
median_income_merge['usa_median_income_latest'] = usa_latest
median_income_merge['usa_median_income_prev'] = usa_previous
```
