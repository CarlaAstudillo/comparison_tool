```python
import pandas as pd
```

```python
#Table B01003

usa_prev = 301461533

usa_latest = 316515021

nj_prev = 8650548

nj_latest = 8904413
```

```python
#Add in county rows
data_latest = pd.read_csv("population_1115.csv")
data_prev = pd.read_csv("population_0509.csv")
```

```python
data_prev.tail(5)
```

```python
data_latest = data_latest[data_latest['GEO.id'] != 'Id']
data_prev = data_prev[data_prev['GEO.id'] != 'Id']
```

```python
data_prev.head(5)
```

```python
data_latest = data_latest.rename(columns = {'GEO.display-label':'GEOdisplaylabel'})
data_prev = data_prev.rename(columns = {'GEO.display-label':'GEOdisplaylabel'})
```

```python
data_latest['GEOdisplaylabel'] = data_latest['GEOdisplaylabel'].replace(to_replace=', New Jersey', value='', regex=True)
data_prev['GEOdisplaylabel'] = data_latest['GEOdisplaylabel'].replace(to_replace=', New Jersey', value='', regex=True)
```

```python
data_prev
```

```python
data_latest_newcols = pd.DataFrame(data_latest.GEOdisplaylabel.str.split(',',1).tolist(), columns = ['GEOdisplaylabel', 'County'])
data_latest = pd.concat([data_latest_newcols, data_latest], axis=1)
```

```python
data_latest['County'] = data_latest['County'].replace(to_replace='County', value='', regex=True)
```

```python
data_latest.columns = ['GEOdisplaylabel', 'County', 'GEO.id', 'GEO.id2', 'GEOdisplaylabel_long', 'population_latest', 'population_latest_moe']
data_latest.head(5)
```

```python
data_prev.dtypes
```

```python
data_prev['GEOdisplaylabel'] = data_prev['GEOdisplaylabel'].astype(str)
data_prev_newcols = pd.DataFrame(data_prev.GEOdisplaylabel.str.split(',',1).tolist(), columns = ['GEOdisplaylabel', 'County'])
data_prev = pd.concat([data_prev_newcols, data_prev], axis=1)
data_prev['County'] = data_prev['County'].replace(to_replace='County', value='', regex=True)
data_prev.columns = ['GEOdisplaylabel', 'County', 'GEO.id', 'GEO.id2', 'GEOdisplaylabel_long', 'population_prev', 'population_prev_moe']
data_prev.head(5)
```

```python
#Merge the Data
data_merge = pd.merge(data_latest, data_prev, left_on='GEO.id2', right_on='GEO.id2', how='left')
data_merge.head(5)
```

```python
len(data_merge.index)
```

```python
data_merge = data_merge.drop(['GEOdisplaylabel_y', 'County_y', 'GEO.id_y', 'GEOdisplaylabel_long_y'], axis=1)
data_merge.rename(columns={'GEOdisplaylabel_x': 'GEOdisplaylabel', 'County_x': 'County', 'GEO.id_x': 'GEO.id', 'GEOdisplaylabel_long_x': 'GEOdisplaylabel_long'}, inplace=True)
```

```python
data_merge.head(5)
```

```python
data_merge.dtypes
```

```python
#Add inflation
#median_income_merge['median_income_prev'] = pd.to_numeric(median_income_merge['median_income_prev_woinf'], errors='coerce') * (252.463/229.343)
#median_income_merge.head(5)
```

```python
#Take out County subdivisions not defined
data_merge = data_merge[data_merge['GEOdisplaylabel'] != "County subdivisions not defined"]
data_merge.head(5)
```

```python
#Find MOE percent
data_merge['population_moe_latest_per'] = (pd.to_numeric(data_merge['population_latest_moe'], errors='coerce')/pd.to_numeric(data_merge['population_latest'], errors='coerce'))*100
data_merge['population_moe_prev_per'] = (pd.to_numeric(data_merge['population_prev_moe'], errors='coerce')/pd.to_numeric(data_merge['population_prev'], errors='coerce'))*100 
```

```python
data_merge.head(4)
```

```python
#Add NJ and Nationwide
data_merge['nj_population_latest'] = nj_latest
data_merge['nj_population_prev'] = nj_prev
data_merge['usa_population_latest'] = usa_latest
data_merge['usa_population_prev'] = usa_prev
```

```python
data_merge['population_change'] = ((data_merge['population_latest']-pd.to_numeric(data_merge['population_prev'], errors='coerce'))/data_merge['population_prev'])*100
data_merge['nj_population_change'] = ((data_merge['nj_population_latest']-pd.to_numeric(data_merge['nj_population_prev']))/data_merge['nj_population_prev'])*100
data_merge['usa_population_change'] = ((data_merge['usa_population_latest']-pd.to_numeric(data_merge['usa_population_prev']))/data_merge['usa_population_prev'])*100
```
