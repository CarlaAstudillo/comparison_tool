```python
import pandas as pd
```

```python
#Add all datasets
#median income is already in census data towns
median_income = pd.read_csv("census_data_towns.csv")
population = pd.read_csv("population_data_towns.csv")
age = pd.read_csv("median-age.csv")
race = pd.read_csv("race-w-hisp.csv")
poverty = pd.read_csv("poverty.csv")
home_values = pd.read_csv("housingvalues_v2.csv")
home_costs = pd.read_csv("housingcosts.csv")
commute = pd.read_csv("commutetimes.csv")
```

```python
len(population.index)
```

```python
len(median_income.index)
```

```python
merge1 = pd.merge(median_income, population, left_on='GEO.id2', right_on='GEO.id2', how='left')
```

```python
len(merge1.index)
```

```python
merge1.head(5)
```

```python
merge1 = merge1.drop(['GEOdisplaylabel_y', 'County_y', 'GEO.id_y', 'GEOdisplaylabel_long_y'], axis=1)
merge1.rename(columns={'GEOdisplaylabel_x': 'GEOdisplaylabel', 'County_x': 'County', 'GEO.id_x': 'GEO.id', 'GEOdisplaylabel_long_x': 'GEOdisplaylabel_long'}, inplace=True)
```
